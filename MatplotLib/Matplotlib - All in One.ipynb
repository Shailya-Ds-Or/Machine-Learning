## Basic Plot using MatplotLib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
x
```

```python
y
```

```python
plt.title("X Vs Y", fontsize=20)
plt.xlabel("X",fontsize=10)
plt.ylabel("Y",fontsize=10)
plt.xlim(0,11)
plt.ylim(0,20)
plt.plot(x,y, color="red", linestyle="--", linewidth=2)

plt.savefig("First_Plot.jpg")
plt.show()

```

## Using figure Object

```python
x = np.linspace(0,10,11)
y = 2*x

a = np.arange(0,10)
b = a**3
```

```python
a
```

```python
b
```

```python
x
```

```python
y
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)

plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(a,b)

plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlabel("a")
axes1.set_ylabel("b")
axes1.set_title("a Vs b")


axes2 = fig.add_axes([0.3,0.3,0.3,0.3])
axes2.plot(x,y)
axes2.set_xlabel("x")
axes2.set_ylabel("y")
axes2.set_title("x Vs y")
```

```python
fig = plt.figure(figsize=(10,8))

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlabel("a")
axes1.set_ylabel("b")
axes1.set_title("a Vs b")


axes2 = fig.add_axes([0.2,0.2,0.3,0.3])
axes2.plot(a,b)
axes2.set_xlabel("x")
axes2.set_ylabel("y")
axes2.set_xlim(2,4)
axes2.set_ylim(0,80)
axes2.set_title("Zoomed In")

fig.savefig("Zoomed In Plot.jpeg",bbox_inches= "tight")
```

## Subplots

```python
a1 = np.arange(0,10)
a2 = a1*2
```

```python
b1 = np.linspace(0,10,11)
b2 = b1**3
```

```python
c1 = np.arange(0,10)
c2 = c1+1.6
```

```python
d1 = np.linspace(0,10,11)
d2 = d1**4
```

```python
a1
```

```python
a2
```

```python
b1
```

```python
b2
```

```python
c1
```

```python
c2
```

```python
d1
```

```python
d2
```

```python
fig,axes = plt.subplots()

axes.plot(a1,a2)

plt.show()
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(a1,a2)
axes[1].plot(b1,b2)

plt.show()
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(a1,a2)
axes[0][1].plot(b1,b2)
axes[1][0].plot(c1,c2)
axes[1][1].plot(d1,d2)

plt.show()
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

axes[0][0].plot(a1,a2,color="red",linestyle="--",linewidth=2)
axes[0][0].set_xlabel("a1")
axes[0][0].set_ylabel("a2")
axes[0][0].set_title("a1 VS a2")

axes[0][1].plot(b1,b2,color="blue",linestyle="-.",linewidth=3)
axes[0][1].set_xlabel("b1")
axes[0][1].set_ylabel("b2")
axes[0][1].set_title("b1 VS b2")

axes[1][0].plot(c1,c2,color="green",linestyle="-",linewidth=0.5)
axes[1][0].set_xlabel("c1")
axes[1][0].set_ylabel("c2")
axes[1][0].set_title("c1 VS c2")

axes[1][1].plot(d1,d2,color="yellow",linestyle="--",linewidth=4)
axes[1][1].set_xlabel("d1")
axes[1][1].set_ylabel("d2")
axes[1][1].set_title("d1 VS d2")

fig.suptitle("SubPlots", fontsize= 30)

plt.tight_layout()

plt.savefig("Subplots.jpg", bbox_inches="tight")
plt.show()
```
