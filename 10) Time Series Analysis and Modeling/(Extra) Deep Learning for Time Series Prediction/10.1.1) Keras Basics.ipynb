# Keras Basics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import keras
```

```python
# Creating Simple Linear Data y = mx + b + noise:

m = 2
b = 3
x = np.linspace(0,50,100)

np.random.seed(101)
noise = np.random.normal(loc= 0, scale= 4, size= len(x))
```

```python
y = m*x + b + noise
```

```python
plt.plot(x, y, "*")
```

### Creating a Neural Network for Line of Best Fit:

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(4, input_dim= 1, activation= "relu"))

model.add(Dense(4, activation= "relu"))

model.add(Dense(1, activation= "linear"))

model.compile(loss= "mse", optimizer= "adam")
```

```python
model.summary()
```

```python
model.fit(x,y, epochs= 500, verbose=1)
```

```python
x_new = np.linspace(0,50,1000)
```

```python
pred = model.predict(x_new)
```

```python
# Plotting Actual and Predicted:

plt.plot(x, y, "*", color= "black")
plt.plot(x_new, pred, color= "red")
```
