### Function for Displaying Clusters:

```python
def display_clusters(model, data):
    
    clusters = model.fit_predict(data)
    
    plt.figure(figsize= (8,4), dpi= 200)
    sns.scatterplot(data= data, x= "X1", y="X2", hue= clusters, palette= "dark")
    plt.show()
```

### K-Means effect on all Datasets:

#### 1) Blobs:

```python
kmc = KMeans(n_clusters= 3)
```

```python
display_clusters(kmc, blobs)
```

#### 2) Moons:

```python
kmc = KMeans(n_clusters= 2)
```

```python
display_clusters(kmc, moons)
```

#### 3) Circles:

```python
kmc = KMeans(n_clusters= 2)
```

```python
display_clusters(kmc, circles)
```

### DBSCAN effect on all Datasets:

#### 1) Blobs:

```python
dbscan = DBSCAN(eps= 0.5)
```

```python
display_clusters(dbscan, blobs)
```

```python
# -1 is not a cluster number here.
# -1 in DBSCAN represents points that are considered Outliers.

# We can Increase or Decrease Outliers by changing "eps" value.
```

```python
dbscan = DBSCAN(eps= 1.5)
```

```python
display_clusters(dbscan, blobs)
```

#### 2) Moons:

```python
dbscan = DBSCAN(eps= 0.2)
```

```python
display_clusters(dbscan, moons)
```

#### 3) Circles:

```python
dbscan = DBSCAN(eps= 0.3)
```

```python
display_clusters(dbscan, circles)
```
